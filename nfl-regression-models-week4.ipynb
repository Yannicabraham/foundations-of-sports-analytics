## Part 2 - Regression Analysis 1 - üèüÔ∏è Test of Home Game Advantage

In this regression analysis, I will try to determine if playing at home gives teams any advantage in a team's performance.
I will run the following regressions where the variable "score_diff" is the dependent variable. 

- (1) Includes a single varialbe "home" as the independent variable
- (2) Includes the "home", "stadium_capacity", "stadium_neutral" variables and the **interaction** between "home" and "stadium_neutral" as the independent variables
- (3) Includes the "home", "stadium_capacity", and "stadium_neutral" variables, the **interaction** between "home" and "stadium_neutral", as well as "team" and "opponent" as the independent variables.

```python
Reg1_1 = sm.ols(formula = 'score_diff ~ home', data=NFL_Game).fit()
print(Reg1_1.summary())
```

## üîé Findings

### Estimated coefficents
- The regression model delivers a positive relationship between both variables, the scoring difference and whether a team plays at home
- playing at home increases the scoring difference by 5.2240 points, holding all other variables constant

### P-Value
- With a p-value of 0.000 we can neglect the null hypothesis. The coefficent is statistically significant at the 1% level.

### R-Squared
- The R-Squared value is relatively low with 0.030. What this basically tells us, is that only 3% of the variation in the scoring difference can be explained by the regression model. It is only a poor fit, other factors play into the dependent variable as well.

```python
Reg1_2 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral', data=NFL_Game).fit()
print(Reg1_2.summary())
```

## üîé Findings

### Estimated coefficents
- The estimated coefficient for the scoring difference slightly increased to 5.2765 compared to the 5.2240 of the previous regression model. It is still statistically significant at the 1% level.
-  the affect of the stadium capacity on the scoring difference is very low with approx. 0.000000005195. It is also to be noted that the coefficent is not statistically significant at the established levels.
-  Playing in a neutral stadium where neither team is at home increases the scoring difference by 7.2518. The coefficient is statiscally significant at the 1% level.
-  The interaction between home and playing in a neutral stadium has a negative impact on the scoring difference with -14.5038. This will likely be a mental component as well, as a team who was meant to play at home now has to waive the impact of home fans, atmosphere and so on. This coeffient is also significant at the 1% level.

### R-Squared
- The R-Squared value only increased slightly to 0.031 compared to the previous regression model. We can not confidently say that this regression model suits as a better fit for the scoring difference. 

```python
Reg1_3 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral+team*opponent', data=NFL_Game).fit()
print(Reg1_3.summary())
```

## Part 3 - Regression Analysis 2 -- ‚ùÑ üå¨Ô∏è Impact of Outside Factors on Scores

In the following analysis, I will investigate if the final score earned by a team is affected by outside factors such as the size and condition of the stadium as well as weather conditions at the stadium.

I will run the following regressions where the variable "score" is the dependent variable.

- (1): Including "season" and "home" variables as independent variables
- (2): Including "season", "home", "weather_temperature", "weather_wind_mph", and "weather_humidity" as independent variables
- (3): Including "season", "home", "weather_temperature", "weather_wind_mph", "weather_humidity", "stadium_capacity", "stadium_age", "stadium_type", "stadium_neutral", and the **interaction** between "home" and "stadium_neutral" as independent variables.
- (4): Including "season", "home", "weather_temperature", "weather_wind_mph", "weather_humidity", "stadium_capacity", "stadium_age", "stadium_type", "stadium_neutral", and the **interaction** between "home" and "stadium_neutral" as independent variables. Addtionally, add both "team" and "opponent" in the regression.

```python
Reg2_1 = sm.ols(formula = 'score ~ season+home', data=NFL_Game).fit()
print(Reg2_1.summary())
```

```python
Reg2_2 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity', data=NFL_Game).fit()
print(Reg2_2.summary())
```

```python
Reg2_3 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity+stadium_capacity+stadium_age+stadium_type+stadium_neutral+home*stadium_neutral', data=NFL_Game).fit()
print(Reg2_3.summary())
```

```python
Reg2_4 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity+stadium_capacity+stadium_age+stadium_type+stadium_neutral+home*stadium_neutral+team*opponent', data=NFL_Game).fit()
print(Reg2_4.summary())
```
